### Problem 1

Consider a  slab of charge of thickness $d$ centered on the $x$ axis. The slab has infinite extent in the $y$ and $z$ direction. The charge density is uniform inside the slab and of magnitude $\rho_{0}$. There is no charge outside of the slab. Give your answers to the following questions in terms of $\rho_{0}$,$d$,$\varepsilon_{0}$ and $x$.

(a) (5 points)  What is the electric field in the $x$ direction for $-\frac{d}{2}<x < \frac{d}{2}$?

You can obtain this choosing different gaussian surfaces. if you choose a cylinder of length 2x centered around x=0 you get:

$E2A=\rho A2x/\epsilon_0\rightarrow E=\frac{\rho x}{\epsilon_0}$

(b) (5 points)  What is the electric field in the $x$ direction  for $x\geq \frac{d}{2}$?

$E=\frac{\rho d}{2\epsilon_0}$


(c) (5 points) What is the electric field for $x\leq -\frac{d}{2}$?

$E=-\frac{\rho d}{2\epsilon_0}$

(d) (5 points)  What is the potential difference between $x=0$ and $x=\frac{d}{2}$?

$\Delta V=-\int_{0}^{\frac{d}{2}}E\,dx=-\frac{1}{8\epsilon_0}\rho_{0}d^{2}$

(e) (5 points) What is the potential difference between $x=-d$ and $x=d$.

 $\Delta V=-\int_{-d}^{d}E\,dx=0 \mathrm{V}$

### Q4

A tightly wound solenoid is made from  wire with a diameter of 0.4 mm.  The cross-sectional area of the solenoid is $2 \,\mathrm{cm^{2}}$ and 12 m of wire was used to make the solenoid

(a) (8 points) If a current of 0.5 A flows through the wire what is the magnitude of the magnetic field inside the solenoid?

 $B=\mu_{0}\frac{N}{l}I=\mu_{0}\frac{1}{0.4\times10^{-3}}\times0.5=1.57\times10^{-3}\mathrm{T}$

(b) (9 points) What is the self-inductance of the solenoid?

 $L=\frac{\mu_{0}N^{2}A}{l}=\mu_0\times\frac{N}{l}NA$

$\pi r^{2}=.0002$

$r=.00798\mathrm{m}$

$N=\frac{12}{2\pi r}=239$

$L=4\pi\times 10^{-7}\frac{1}{0.4\times10^{-3}}\times 239\times0.0002=1.57\times10^{-4}\mathrm{H}$

(c) (8 points) The solenoid is disconnected from the power supply. It is then placed in a uniform magnetic field of strength 0.2 T which runs down it's axis. Someone then squashes it with a hammer so that in 0.04 s it's area is reduced to (approximately) zero. What is the average emf generated across the entire solenoid during the 0.04 s that the solenoid is being squashed?

 $\mathcal{E}=N\frac{d\Phi_{B}}{dt}=N\frac{dA}{dt}B=239\frac{0.0002}{0.04}0.2=0.24\mathrm{V}$


### Q5

A crab hunter is hunting crabs on a liquid methane sea on Titan (a moon of Saturn). The depth of the liquid methane is 20m and it's refractive index is $n=1.27$. Unfortunately the crab **from space** that he is hunting has his own laser defense system. The crab can adjust the angle of his laser shot so that he will hit the hunter who is 1m above the surface. 

(a) (4 points) How deep does the liquid methane sea appear to be to the hunter (looking straight down)?

 $d'=\frac{d}{n}=\frac{20}{1.27}=15.75\mathrm{m}$

(b) (3 points) If the wavelength of the laser light in air is 650nm, what is it's wavelength in liquid methane?

$\frac{650}{1.127}=511\mathrm{nm}$

(c) (3 points) What is the speed of the light in the liquid methane?

 $\frac{3\times10^{8}}{1.27}=2.36\times10^{8}\mathrm{ms^{-1}}$


(d) (3 points) What is the frequency of the light in liquid methane?

 $v=f\lambda$

$f=\frac{v}{\lambda}=4.62\times10^{14}\mathrm{Hz}$

(e) (4 points) The crab can only shoot the hunter if total internal reflection does not occur. What is the maximum value of the angle $\theta$ that the crab can shoot at and hit the hunter?

$\sin(\theta_{c})=\frac{1}{n}$

$\theta_{c}=51.94^{o}$


(f) (8 points) Rather than shooting at the angle you found in part (e), the crab waits until the hunter is a total of 2m horizontal distance (as shown in the diagram) and then fires his laser for a direct hit. What is the angle $\theta$ at which he fires his laser. You may use small angle approximations in this problem.

 $2=20\tan\theta_{i}+\tan\theta_{r}\approx 20\tan\theta_{i}+\theta_{r}$

$\theta_{r}\approx 1.27\theta_{i}$

$2=21.27\theta_{i}$

$\theta_{i}=\frac{2}{21.27}=0.094 \mathrm{rads}=5.39^{o}$
